```python
# ==================
# IMPORTS LIBRARIES
# =================

# Data Handling
import pandas as pd
import numpy as np

# Data Visualization
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# ==============
# Load Dataset
# ==============

df = pd.read_csv('amazon.csv')
df.head(5)
```

### Data Cleaning

```python
# Drop Unnecessary Columns to kee  

df = df.drop(columns=['user_id','user_name','review_id','review_title','review_content','img_link','product_link','about_product'])
```

```python
# -------------------------------
# Data Type Conversion & Cleaning
# -------------------------------


# Cleaning and Converting Discounted_price and Actual_price
price_cols = ['discounted_price','actual_price'] 
for col in price_cols:
    df[col] = ( 
    df[col].astype(str)
    .str.strip()
    .str.replace(r'[^\d.]','',regex=True)
    )
    # Convert to Numeric 
    df[col] = pd.to_numeric(df[col],errors='coerce')

# Cleaning Discounted_percentage column
df['discount_percentage'] = (
    df['discount_percentage'].astype(str).str.strip()
    .str.replace('%','',regex=True)
)
# convert to Numeric
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

# Cleaning and Converting rating and rating_count cols 
rating_cols = ['rating','rating_count'] 
for col in rating_cols:
    df[col] = ( 
        df[col].astype(str)
        .str.replace(',','',regex=True)
        .str.strip()
    )
    df[col] = pd.to_numeric(df[col],errors='coerce')
```

During preprocessing, incorrect initial conversions resulted in loss of data. The dataset was reloaded and a more robust cleaning approach was applied to accurately extract numeric values from price columns.

Data must be clean before conversion,
Wrong regex = broken data,
errors='coerce' silently replaces invalid values

```python
# ========================
# Reloads Problematic Cols
# ========================

# Reloads problematic Cols ['discounted_pricec','actual_price'] again from the csv cause the existing one is broken (NAN).

reload_col = ['discounted_price','actual_price'] 
df_prices = pd.read_csv('amazon.csv', usecols=reload_col)

# Replace the broken columns in your Dataframe with New One 
df[['discounted_price','actual_price']] = df_prices
```

```python
# Verify Reloaded Columns

df[['discounted_price','actual_price']].head(10)
```

```python
# Check for unique values / hidden issues
for col in ['discounted_price', 'actual_price']:
    print(col, df[col].unique()[:10])
```

```python
# Cleaning Prices Columns
price_cols = ['discounted_price', 'actual_price']

for col in price_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace('₹', '', regex=False)   # remove ₹ symbol
        .str.replace(',', '', regex=False)  # remove comma
        .str.strip()
    )
    
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.info()
```

After cleaning and transforming the dataset, all numerical columns were successfully converted into appropriate data types. Minor missing values were handled to ensure data quality for analysis.

```python
# ========================
# Handling Missing Values 
# =======================

missing_values = df.isnull().sum().sort_values(ascending=False)
missing_values 
```

```python
# Drops Missing Values 
df.dropna(subset=['rating_count','rating'],inplace=True)
```

### Feature Engineering 

```python
# Extracts New Feature from rating_count and rating 
df['popularity_score'] = (df['rating_count'] * df['rating'])
```

```python
# Extract main category
# Overall Category column have very long name, let's take the main name of the category
df['main_category'] = df['category'].apply(lambda x: x.split('|')[0])
```

```python
# As Category col, the Product col also have long names. 
# Extract First four words of the product name 
df['product_name_short'] = df['product_name'].apply(lambda x: ' '.join(x.split()[:4]))
```

```python
df['product_name_short'].unique()
```

## Exploratory Data Analysis (EDA)

### Category-Level Insights

```python
# Which category has the highest average rating?

high_category_rating = df.groupby('main_category')['rating'].mean().head(5).sort_values(ascending=False)
sns.barplot(x=high_category_rating.values,y=high_category_rating.index,color='orange')
plt.title('Category With Highest Average Rating')
plt.xticks(rotation=90)
plt.xlabel('Rating')
plt.ylabel('Categories')
```

```python
# Which Category has highest Discount?

category_discount = df.groupby('main_category')['discount_percentage'].mean().sort_values(ascending=False)
top_5 = category_discount.head(5)

sns.barplot(x=top_5.values,y=top_5.index,palette='coolwarm')
plt.title('Category With Highest Discount')
plt.xticks(rotation=60)
plt.xlabel('Discount Percentage')
plt.ylabel('Categories')
```

```python
# Which Category has the Most Popularity?

popular_category = df.groupby('main_category')['popularity_score'].mean()
top = popular_category.head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top.index,y=top.values)
plt.title('Categories Vs Popularity')
plt.xlabel('Popularity')
plt.xlabel('Categories')
plt.xticks(rotation=35)
plt.grid(True)
```

### Product-based Analysis 

```python
# Which products are most popular ?

popular_products = df.groupby('product_name_short')['popularity_score'].mean().sort_values(ascending=False)
top_10 = popular_products.head(10)
plt.figure(figsize=(14,6))
sns.barplot(x=top_10.index,y=top_10.values,color='orange')
plt.title('Top-10 Popular Products')
plt.xlabel('Products')
plt.ylabel('Popularity (In-Millions) ')
plt.xticks(rotation=45)
```

```python
# top-5 product highest ratings?

highest_rating_product = df.groupby('product_name_short')['rating'].mean()
top_5 = highest_rating_product.head(5)
plt.figure(figsize=(9,7))
sns.barplot(x=top_5.values,y=top_5.index)
plt.title('top-5 Products Highest Rating')
plt.xlabel('Rating')
plt.ylabel('Products')
plt.xticks(rotation=45)
```

```python
#  High rating Products but low popularity and Low Rating Products but with high Popularity?

# Scatter Plot to show Relationship

sns.relplot(data=df,x='popularity_score',y='rating',kind='scatter')
plt.title('Products With High Rating But Low Popularity')
plt.xlabel('Popularity (Millions) ')
plt.ylabel('Rating')
```

**High rating,but Low Popularity**

A lot of products are very good but not discovered yet. These are the best opportunity products (need Promotion).

**Low rating , but High Popularity**

There is no strong evidence of products that are widely popular but poorly rated.

## Pricing Analysis

```python
# Price Distribution 

plt.figure(figsize=(10,5))
sns.histplot(df['discounted_price'], bins=50)
plt.title("Distribution of Discounted Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
# Prices vs Rating?

plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x='discounted_price', y='rating')
plt.title("Price vs Rating")
plt.xlabel("Discounted Price")
plt.ylabel("Rating")
plt.show()
```

There is Week Relationship between Rating and Discounted Price. Price does not strongly influence customer ratings.

```python
# Range Analysis of Pricing 

# Define meaningful bins
bins = [0, 10000, 20000, 50000, 80000]
labels = ['0–10k', '10k–20k', '20k–50k', '50k–80k']
df['price_range'] = pd.cut(df['actual_price'], bins=bins, labels=labels)

# Count products in each range
counts = df['price_range'].value_counts().sort_index()
plt.figure()
counts.plot(kind='bar')
plt.title("Products in Different Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Number of Products")
plt.xticks(rotation=0)
plt.show()
```

## Correlation Analysis

```python
# heatmap (Correlation Matrix)
plt.figure(figsize=(8,6))
sns.heatmap(df[['discounted_price','actual_price','discount_percentage','rating','rating_count','popularity_score']].corr(), 
            annot=True)
plt.title("Correlation Matrix")
plt.show()
```

Discounted and actual prices are nearly identical in behavior (r=0.96), while higher discount percentages are linked to lower-priced and lower-rated products. Price has virtually no impact on popularity, and rating count perfectly mirrors popularity score, making one of them redundant.
